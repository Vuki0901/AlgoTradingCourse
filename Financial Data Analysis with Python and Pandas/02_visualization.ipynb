```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8")
```

```python
df = pd.read_csv("multi_assets.csv", header=[0, 1], index_col=0, parse_dates=[0])
df
```

```python
df.info()
```

```python
df.Close
df.Close.BA
df.loc["2023-10-05"]
```

```python
close = df.Close.copy()
close
```

```python
close.describe()
```

```python
close["^DJI"].dropna().plot(figsize = (15, 8), fontsize = 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
close["MSFT"].div(close["MSFT"][0]).mul(100)
```

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.dropna().plot(figsize=(15, 8), fontsize=13, logy=True)
plt.legend(fontsize = 13)
plt.show
```
