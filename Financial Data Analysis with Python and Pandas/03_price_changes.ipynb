```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8")
```

```python
close = pd.read_csv("close.csv", index_col="Date", parse_dates=[0])
close
```

```python
msft = close.MSFT.dropna().to_frame().copy()
msft
```

```python
msft.rename(columns={"MSFT": "Price"}, inplace=True)
msft
```

```python
msft.shift(periods=1)
```

```python
msft["P_lag1"] = msft.shift(periods = 1)
msft
```

```python
msft["P_diff"] = msft.Price.sub(msft["P_lag1"])
msft
```

```python
msft["P_diff2"] = msft.Price.diff(periods=1)
msft.P_diff.equals(msft.P_diff2)
```
